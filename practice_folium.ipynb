```python
# 마커 그리기
map = folium.Map(location=[37.505, 126.77], zoom_start=16, tiles='stamenwatercolor')
tooltip = '클릭해 주세요!'
folium.Marker([37.510, 126.78], popup='<i>위</i>', tooltip=tooltip).add_to(map)
folium.Marker([37.505, 126.78], popup='가운데', tooltip=tooltip).add_to(map)
folium.Marker([37.50, 126.78], popup='<b>아래</b>', tooltip=tooltip).add_to(map)
map
```

```python
map2 = folium.Map([37.5, 126.78], zoom_start=16)
folium.Marker([37.510, 126.78], icon=folium.Icon(icon='cloud')).add_to(map2)
folium.Marker([37.505, 126.78], icon=folium.Icon(icon='star', color='yellow')).add_to(map2)
folium.Marker([37.50, 126.78], icon=folium.Icon(icon='bookmark', color='green')).add_to(map2)
map2
```

```python
folium.Circle([37.5, 126.773], radius=100, color='#ABF200', fill=True, fill_color='#AA0000').add_to(map2)
folium.Circle([37.51, 126.773], radius=200, color='#AAAA00', fill=True, fill_color='#AA0000').add_to(map2)
map2
```

```python
map2.add_child(folium.ClickForMarker(popup='Waarning'))
```

```python
map2.add_child(folium.LatLngPopup())
```
