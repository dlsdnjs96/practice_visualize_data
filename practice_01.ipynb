```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.set(style='darkgrid')
tips = sns.load_dataset('tips')
tips
```

# relplot (relation plot)

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
# hue = 색상
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='size')
```

```python
# style & hue
sns.relplot(x='total_bill', y='tip', data=tips, style='smoker', hue='size')
```

```python
# color
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', palette='Blues_d')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, size='size', hue='size', palette='Blues_d')
```

```python
df = pd.DataFrame(dict(time=np.arange(500), value=np.random.randn(500).cumsum()))
df
```

```python
sns.relplot(x='time', y='value', data=df, kind='line')
```

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.relplot(data=fmri, x='timepoint', y='signal', kind='line', ci=None)
sns.relplot(data=fmri, x='timepoint', y='signal', kind='line', ci='sd')
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, hue='region', style='event', kind='line')
```

# catplot (category plot)

```python
sns.catplot(data=tips, x='day', y='total_bill')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='swarm')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='box')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='violin', style='sex')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='boxen')
```

```python
sns.catplot(data=tips, x='day', kind='count')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='bar')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='point')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='swarm', hue='sex')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='violin', split=True, hue='sex')
```
